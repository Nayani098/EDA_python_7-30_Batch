```python
# first create 4 individual function
# fun1: add
# fun2: sub
# fun3: mul
# fun4: div

# create a main funtion name: calculator
# inside main function
# print some statements
# option1: addition
# option2: sub, opt3: mul, opt4: div
# option=eval(input)
# a value
# b value
# if option==1:
#    add function
# .....

def add(a,b):
    print(a+b)
def sub(a,b):
    print(a-b)
def mul(a,b):
    print(a*b)
def div(a,b):
    print(a/b)

def calculator():
    print('option1: add')
    print('option2: sub')
    print('option3: mul')
    print('option4: div')
    option=eval(input('enter one option (1-4):'))
    a=eval(input('enter a value:'))
    b=eval(input('enter b value:'))
    if option==1:
        add(a,b)
    elif option==2:
        sub(a,b)
    elif option==3:
        mul(a,b)
    elif option==4:
        div(a,b)
    else:
        print('enter correct option')
calculator()
```
